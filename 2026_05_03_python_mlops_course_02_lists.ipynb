```python
#iterating over lists

colors = ["red", "green", "blue"]

for i in colors:
    print(i)
```

```python
for color in colors:
    print(color)
```

```python
nombre = "Carlos"
for letra in nombre:
    i = nombre.index(letra) + 1
    print(f"La letra numero {i} es :{letra}" )
```
