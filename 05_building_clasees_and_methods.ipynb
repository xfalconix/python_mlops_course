```python
#State:
 
class Animal:
    
    def __init__(self, name, legs=4, barks=True):
        self.name = name
        self.legs = legs
        self.barks = barks
    
    def info(self):
        print(f"This is an animal named {self.name}, has {self.legs} legs")
        if self.barks:
            print("And this barks.")
        else:
            print("It does not bark at all.")

```

```python
bunny = Animal("Bunny", barks=False)
bunny.info()
```

```python
print("Bunny's name:", bunny.name)
print("Bunny's legs:", bunny.legs)
print("Does Bunny bark?", bunny.barks)
```

```python
#Adding methods  

#class with two methods or more
#presupuesto, gasto, agregar fondos
class Presupuesto:
    def __init__(self, total):
        self.total = total

    def gastar(self, cantidad):
        if cantidad > self.total:
            print(f"No tienes suficientes fondos. Tu saldo total es: {self.total}")
        else:
            self.total = self.total - cantidad
            print(f"Has gastado {cantidad}. Te quedan {self.total}.")
    
    def agregar(self, cantidad):
        self.total = self.total + cantidad
        print(f"Has agregado {cantidad}. Tu saldo total es: {self.total}.")
```

```python
#a method can call another method
class Budget:
    def __init__(self, budget):
        self.budget = budget

    def expense(self, amount):
        self.budget = self.budget - amount
        self.report()
    
    def report(self):
        print(f"Your remaining budget is: {self.budget}")
```

```python
april_budget = Budget(100)
april_budget.expense(30)
```

```python
april_budget.budget
```

```python
april_budget.budget
```

```python
class Budget:
    def __init__(self, budget):
        self.budget = budget

    def expense(self, amount):
        self.budget = self.budget - amount
        self.report()
    
    def report(self, currency="$"):
        print(f"Your remaining budget is: {currency}{self.budget}")
```

```python
may_budget = Budget(200)
may_budget.report()
```

```python
may_budget.report("€")
```

```python
#Class Inheritance
#A parent class and a child class that inherits from the parent class
#Usefull when you want to create a new class that is a modified version of an existing class. You can create a new class that inherits the attributes and methods of the existing class, and then add or modify the attributes and methods as needed.

#Create a base class for house pets

class Pet:

    def eat(self):
        self.food = self.food - self.appetite
        print(f"Ate {self.appetite} units of food. Remaining food: {self.food}")
```

```python
#create a child classes for other house pets like cat and dog and parakeet

class Parakeet(Pet):

    def __init__(self):
        self.food = 100
        self.appetite = 1

class Dog(Pet):

    def __init__(self):
        self.food = 500
        self.appetite = 10


```

```python
perry = Parakeet()
rufus = Dog()
perry.eat()
rufus.eat()
```

```python
# demonstrate how other classes have methods that automatically appear
for attribute in dir(rufus):
    if attribute.startswith("__") == False:
        print(attribute)
```
