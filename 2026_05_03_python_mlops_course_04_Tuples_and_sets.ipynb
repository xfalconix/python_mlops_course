```python
#uso de .pop
# Ejemplo con una lista

# Creamos una lista de frutas
frutas = ["manzana", "banana", "cereza", "dátil"]
print(f"Lista original: {frutas}")

# Usamos .pop() sin argumentos: quita el último elemento
ultima_fruta = frutas.pop()
print(f"Fruta que quité (con .pop()): {ultima_fruta}")
print(f"Lista después de .pop(): {frutas}")

# Usamos .pop() con un índice: quita el elemento en la posición 0 (el primero)
primera_fruta = frutas.pop(0)
print(f"Fruta que quité (con .pop(0)): {primera_fruta}")
print(f"Lista final: {frutas}")


# Ejemplo con un diccionario (requiere una clave)

# Creamos un diccionario de colores favoritos
colores = {"Juan": "azul", "Ana": "verde", "Pedro": "rojo"}
print(f"\nDiccionario original: {colores}")

# Usamos .pop() con una clave: quita la entrada de 'Ana'
color_ana = colores.pop("Ana")
print(f"Color de Ana que quité: {color_ana}")
print(f"Diccionario después de quitar a Ana: {colores}")
```
