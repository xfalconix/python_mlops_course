### Uso de `elif`
`elif` se usa cuando tienes más de dos posibles resultados y necesitas verificar múltiples condiciones en secuencia.

```python
temperatura = 25

if temperatura > 30:
    print("Hace mucho calor.")
elif temperatura > 20:
    print("Hace una temperatura agradable.")
elif temperatura > 10:
    print("Hace fresco.")
else:
    print("Hace frío.")
```

En este ejemplo:
- Primero, se verifica si `temperatura > 30`. Si es `True`, se ejecuta ese bloque y el resto se ignora.
- Si la primera condición es `False`, entonces se verifica la condición del primer `elif`: `temperatura > 20`.
- Si esa también es `False`, se pasa al siguiente `elif`: `temperatura > 10`.
- Si todas las condiciones `if` y `elif` son `False`, entonces se ejecuta el bloque `else`.

Esto asegura que solo se ejecute un bloque de código entre todas las condiciones, y en el orden en que se presentan.

### ¿Por qué `elif` en lugar de solo `if`?

Cuando usas una secuencia de `if` sin `elif`, Python evalúa *cada* condición `if` de forma independiente. Esto significa que si una condición es verdadera y se ejecuta su bloque, el programa *continuará* evaluando las siguientes condiciones `if`.

En contraste, con `if-elif-else`, una vez que se encuentra una condición verdadera (`if` o `elif`) y se ejecuta su bloque de código, Python **salta** el resto de la cadena `elif` y `else`. Esto asegura que solo se ejecute uno de los bloques de código, lo cual es ideal cuando las condiciones son mutuamente excluyentes (es decir, solo una puede ser verdadera a la vez).

```python
# Ejemplo usando solo 'if'
temperatura = 25

print("--- Usando solo 'if' ---")
if temperatura > 30:
    print("Hace mucho calor.")
if temperatura > 20:
    print("Hace una temperatura agradable.")
if temperatura > 10:
    print("Hace fresco.")
else:
    # Este 'else' solo se asocia con el último 'if'
    print("Hace frío.")
```
