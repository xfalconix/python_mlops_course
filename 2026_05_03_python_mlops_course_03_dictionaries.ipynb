```python
#Looping over dictionaries
contact_information = {
    "name": "John",
    "last_name": "Doe",
    "age": 42,
    "phone": "123456789"
}
```

```python
for key in contact_information:
    print(key)
```

```python
for key in contact_information.keys():
    print(key)
```

```python
for value in contact_information.values():
    print(value)
```

```python
for key, value in contact_information.items():
    print(f"{key} ---> {value}")
```
