```python
#Functions basics

#Defining a function with no parameters / arguments
def simple():
    print("This is a simple function.")
simple()
```

```python
simple()
```

```python
#watch out for functions that don't return anything, they return None
result = simple()
print("The value of Result:", result)   
```

```python
#if you want a function to return something, use the return keyword
def produce():
    return "This is a return value!"
result = produce()
print("The value of Result:", result)
```

```python
#an optional argument uses keyword arguments
def greetings(full_name="John Doe"):
    print("Greetings ", full_name)
```

```python
#now you can call the function with or without an argument
greetings()
greetings("Alice")
```

```python
def saludos(nombre="Nombre por defecto"):
    print("Saludos", nombre)
saludos()
saludos("Carlos")
```

```python
def formal_greeting(name, title="Mr./Ms."):
    print(f"Hello {title} {name}!")
formal_greeting("Smith")
formal_greeting("Smith", "Dr.")
```

```python
#Variable arguments and keyword arguments

#this function can take o or more arguments
def family_members(*args):
    for name in args:
        print(name)

family_members("Alice", "Bob", "Charlie")
```

```python
def cars_numbers(*args):
    for car in args:
        print(f"car number {1 + args.index(car)}: {car}")

cars_numbers("Toyota", "Honda", "Ford")
```

```python
#Variable keyword arguments

def stats(**kwargs):
    #kwargs is a dictionary of the keyword arguments passed to the function
    for key, value in kwargs.items():
        print(f"{key} ---> {value}")

stats(mean=5, median=3, mode=2)
```

```python
result = stats()
print("The value of Result:", result)
```
